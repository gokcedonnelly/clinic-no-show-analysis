# Milestone 3 — Data Cleaning

**Dataset:** clinic_appointments_en.csv  
**Goal:** Inspect and clean the data so it’s analysis-ready.

```python
df_clean = df.copy() 
print(df_clean.shape)
df_clean.head()
```

**Check missing + duplicates**

```python
print("Missing values per column:")
print(df_clean.isna().sum())

print("Duplicate rows:", df_clean.duplicated().sum())
```


**Standardize formats**

```python
df_clean["date"] = pd.to_datetime(df_clean["date"], errors="coerce")

gmap = {
    "male": "Male","Male":"Male","MALE":"Male","m":"Male","M":"Male",
    "female":"Female","Female":"Female","FEMALE":"Female","f":"Female","F":"Female"
}
df_clean["gender"] = df_clean["gender"].astype(str).str.strip().str.title().replace(gmap)
df_clean.loc[~df_clean["gender"].isin(["Male","Female"]), "gender"] = "Other"

binary_cols  = ["weekend","sms_reminder","rain","no_show"]
numeric_cols = ["age","distance_km","lead_time_days","prior_no_shows"]

for c in binary_cols:
    df_clean[c] = pd.to_numeric(df_clean[c], errors="coerce").fillna(0).astype(int).clip(0,1)

for c in numeric_cols:
    df_clean[c] = pd.to_numeric(df_clean[c], errors="coerce").fillna(0)
```

**Validity checks**

```python
print("Age valid:", (df_clean["age"].between(0,100)).all())
print("Lead time nonnegative:", (df_clean["lead_time_days"] >= 0).all())
print("Distance nonnegative:", (df_clean["distance_km"] >= 0).all())
print("Prior no_shows nonnegative:", (df_clean["prior_no_shows"] >= 0).all())

for c in binary_cols:
    print(f"{c} unique values:", df_clean[c].unique())
```

**Outlier handling**

```python
def iqr_clip(s: pd.Series) -> pd.Series:
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return s.clip(lower, upper)

for c in numeric_cols:
    df_clean[c] = iqr_clip(df_clean[c])

df_clean[numeric_cols].describe()
```

**final check**

```python
df_clean = df_clean.drop_duplicates()

df_clean.info()
df_clean.head()
```

### Milestone 3 — Cleaning Summary
- Parsed date to datetime for time-based analysis.
- Standardized gender to {Male, Female, Other} to remove inconsistent labels.
- Ensured binary flags (weekend, sms_reminder, rain, no_show) are integers in {0,1}.
- Converted numeric columns (age, distance_km, lead_time_days, prior_no_shows) to numeric and handled any non-numeric values.
- Applied IQR winsorizing to numeric columns to cap extreme outliers (keeps rows while limiting undue influence).
- Validated simple rules (age ≤ 100; nonnegative counts/distances; binaries ∈ {0,1}) and checked/removed duplicates.
The cleaned data below is used for analysis in Milestone 4.

# Milestone 4 — Analysis & Visualization

```python
import matplotlib.pyplot as plt

no_show_rate = df_clean["no_show"].mean()
print("Overall no-show rate: {:.1f}%".format(no_show_rate*100))

df_clean["no_show"].value_counts().plot(kind="bar")
plt.title("Show vs No-Show Counts")
plt.xlabel("No-Show (0=Showed, 1=No-Show)")
plt.ylabel("Count")
plt.show()
```

```python
sms_group = df_clean.groupby("sms_reminder")["no_show"].mean()
print(sms_group)

sms_group.plot(kind="bar")
plt.title("No-Show Rate by SMS Reminder")
plt.xlabel("SMS Reminder (0=No, 1=Yes)")
plt.ylabel("No-Show Rate")
plt.show()
```

```python
df_clean["lead_time_bucket"] = pd.cut(df_clean["lead_time_days"],
                                      bins=[-1,0,3,7,14,30],
                                      labels=["0","1-3","4-7","8-14","15-30"])

lead_group = df_clean.groupby("lead_time_bucket")["no_show"].mean()
print(lead_group)

lead_group.plot(kind="bar")
plt.title("No-Show Rate by Lead Time")
plt.xlabel("Days Until Appointment")
plt.ylabel("No-Show Rate")
plt.show()
```

```python
gender_group = df_clean.groupby("gender")["no_show"].mean()
print(gender_group)

gender_group.plot(kind="bar")
plt.title("No-Show Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("No-Show Rate")
plt.show()
```

### Milestone 4 — Findings
- **Overall no-show rate:** ~14.6%.
- **SMS reminders:** Appointments with an SMS reminder have a lower no-show rate than those without.
- **Lead time:** No-show rate tends to increase for longer wait times (8–14 and 15–30 day buckets are higher).
- **Gender:** Small differences by gender; effect size appears modest relative to SMS/lead time.
These results suggest operational levers like sending reminders and reducing long wait times could reduce no-shows.
